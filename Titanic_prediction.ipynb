```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
urltrain = '/content/train.csv'
urltest = '/content/test.csv'
```

```python
dftrain = pd.read_csv(urltrain)
dftest = pd.read_csv(urltest)
```

```python
submit = dftest
```

```python
dftrain = dftrain.dropna(axis=0,subset=['Embarked'])
```

```python
dftrain.isna().sum()
```

```python
dftrain  = dftrain.drop('Cabin',axis=1)
```

```python
dftrain['Age'].fillna(dftrain['Age'].mean(),inplace=True)
```

```python
dftrain.head(10)
```

```python
dftrain = dftrain.drop(['PassengerId','Name','Ticket','Fare','Embarked'],axis=1)
dftrain['Sex'] = dftrain['Sex'].replace(['male','female'],[1,0])
```

```python
dftrain.head()
```

```python
dftrain = dftrain.drop(['Parch'],axis=1)
```

```python
X_train = dftrain.drop(['Survived'],axis=1)
Y_train = dftrain['Survived']
```

```python
dftest  = dftest.drop('Cabin',axis=1)
dftest = dftest.drop(['PassengerId','Name','Ticket','Fare','Embarked','Parch'],axis=1)
dftest['Sex'] = dftest['Sex'].replace(['male','female'],[1,0])
```

```python
dftest.head()
```

```python
dftest['Age'].fillna(dftest['Age'].mean(),inplace=True)
```

```python
dftest.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
```

```python
X_test = dftest
```

```python
X_test = StandardScaler().fit_transform(X_test)
```

```python
def predmodel(X_train,Y_train):
  from sklearn.linear_model import LogisticRegression
  log = LogisticRegression(random_state=0)
  log.fit(X_train,Y_train)
  print("Training accuracy : ",log.score(X_train,Y_train))
  return log
```

```python
pred = predmodel(X_train,Y_train)
```

```python
prediction = pred.predict(X_test)
```

```python
prediction
```

```python
predict=pd.DataFrame(prediction,columns=['Survived'])
predict
```

```python
submission=pd.concat([submit['PassengerId'],predict],axis=1)
```

```python
submission
```

```python
submission.to_csv('submission.csv',index=False)
```
